```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#import dataframe
crime_df = pd.read_excel('..\data\raw\Crime_statistics_2023V1.xlsx', skiprows=1)
gdp_df = pd.read_excel('..\data\raw\P_Data_Extract_From_World_Development_Indicators.xlsx')
unempl_df = pd.read_excel('..\data\raw\UNEMPLOYMENT.xlsx')
```

```python
#check data
print(crime_df)
```

```python
#reset the index
crime_df.reset_index(drop=True, inplace=True)
print(crime_df)
```

```python
#rename column
crime_df.rename(columns={'Row Labels': 'Country'}, inplace=True)
crime_df.rename(columns={'Grand Total': 'Total Crimes'}, inplace=True)

print(crime_df)
```

```python
#check data
print(gdp_df)
```

```python
#remove useless columns
gdp_df.drop(['Series Name', 'Series Code', 'Country Code'], axis=1, inplace=True)
#rename column
gdp_df.rename(columns={'2021 [YR2021]': 'GDP'}, inplace=True)
gdp_df.rename(columns={'Country Name': 'Country'}, inplace=True)


print(gdp_df)
```

```python
#check df
print(unempl_df)
```

```python
#rename columns
unempl_df.rename(columns={'TIME': 'Country'}, inplace=True)
unempl_df.rename(columns={'2021': 'Unemployment Rate'}, inplace=True)
#delete second row
rows_to_delete = [0, 1, 2, 3]
unempl_df = unempl_df.drop(rows_to_delete)
# Reset the index
unempl_df = unempl_df.reset_index(drop=True)
#delete columns
columns_to_delete = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2022']
unempl_df = unempl_df.drop(columns=columns_to_delete)
#delete empty columns
unempl_df = unempl_df.dropna(axis=1)
```

```python
#check df
print(unempl_df)
```

```python
#merge dataframe
merged_df = pd.merge(crime_df, unempl_df, on='Country')
df = pd.merge(merged_df, gdp_df, on='Country')

print(df)
```

```python
#save in new csv
df.to_csv('..\data\cleaned\data_cleaned.csv', index=False)
```
